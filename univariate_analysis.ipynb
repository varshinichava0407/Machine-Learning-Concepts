```python
df_new[['math score','reading score','writing score']].describe()
```

```python
#count --- how many values exists in that column
#mean - Average Score
#Std(Standard deviation) - Measures Spread - How much scores vary from the mean
#min - smallest value
#25% (Q1) : First Quartile - 25% of students scored below this
# 50% (Median/Q2) - Middle Score(50% above / 50% below)
#75%(Q3) - Third Quartile - 75% scored below this
#max : Highest Value
#Extra Interpretation: mean and median are close, data is too symmetric
#If mean > median, scores are right-skewed (more low score)
#If mean<median , scores are left skewed (more high scores)
#IQR IS 75% - 25%
```

```python
sns.distplot(df_new['math score'],kde=True)
```

```python
#Skewness shows whether the data is clustered more towards high or low values
#Positive Skew(>0): Tail on the right side (more low values, few high outliers) : Mean > Median
#Negative Skew(<0): Tail on the left side(more high values, few low outliers): Mean < Median
df_new['math score'].skew()
```
